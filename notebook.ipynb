```python
import json
import csv
import pandas as pd
import numpy as np
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
#video_id='erx9czQsY2Q'
video_id='kJIgq5n7WDc'
data=YouTubeTranscriptApi.get_transcript(video_id,languages=['en'])
#print (data)
print ("start ="+(data[0]['text']))
print ((data[107]['start']))
print ((data[107]['duration']))
print (data[107]['start']+data[107]['duration'])
print (len(data))
#parsed_data=json.load(data)
#print (parsed_data[0])
```

```python
import re

csvfilename=video_id+".csv"
for i in range(len(data)):
     with open (csvfilename,'a',encoding="utf-8") as res:        
        writer=csv.writer(res)           
        s="{},{},{},{}\n".format(data[i]['duration'],data[i]['start'],data[i]['start']+data[i]['duration'], re.sub(r'\s+', ' ',re.sub(r'\W', ' ',data[i]['text'],flags=re.I)))
        res.write(s)
        print (s)
```

```python
df=pd.read_csv(csvfilename)
df.columns=['duration','start','end','text']
X=df['text']
df
```

```python
from sklearn.externals import joblib
```

```python
emotion=[]
loaded_model = joblib.load('emotion_model.sav')
for i in X:
    emotion.append(loaded_model.predict([i])[0])
df['emotion']=emotion
df    
```

```python
dict={0:"hate_speech",1:"offensive_speech",2:"neither"}
nature=[]
loaded_model = joblib.load('offense_hate_modelv1.sav')
for i in X:
    nature.append(dict[loaded_model.predict([i])[0]])
df['nature']=nature
df    
```
